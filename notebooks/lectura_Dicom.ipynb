## Lectura imagen Dicom

```python
import pydicom
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
%matplotlib inline
```

```python
#Ruta de la imagen DICOM
dicom_path = '/home/juanito/projects/uv-projects/Proyecto_Neumonia/UAO-Neumonia/tests/DICOM/normal (2).dcm'

#Cargamos la imagen DICOM
dicom_img = pydicom.dcmread(dicom_path)

# Mostrar algunos metadatos
print(dicom_img)
```

```python
# Extraemos los pxeles de la imagen
img_array = dicom_img.pixel_array

# Mostramos la imagen
plt.imshow(img_array, cmap='gray')
plt.title("Imagen DICOM")
plt.axis("off")
plt.show()
```
